# Single layered Perceptron

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
dataset = pd.read_csv(r'D:\Academics\Project Data Set\cgpa_score_placed\cgpa_score_placed.csv')
dataset.head(3)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='cgpa', y="resume_score", data = dataset, hue = 'placed')
plt.show()
```

```python
x = dataset.iloc[:,:-1]
y = dataset['placed']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=10)
```

## Modelbuilding: Perceptron 

```python
from sklearn.linear_model import Perceptron

pr = Perceptron()
pr.fit(x_train, y_train)
```

```python
pr.score(x_train, y_train)*100 , pr.score(x_test, y_test)*100
```

```python
plt.figure(figsize=(6,4))
plot_decision_regions(x.to_numpy(), y.to_numpy(),clf=pr)
plt.show()
```

#### We can change Hyperparameters and see if there is any betterment in the result but ***<u>Single Layered Perceptron</u>*** doesn't give better result in such scenrions.
### That's why we use ***<u>Multi Layered Perceptron</u>*** or called as ***<u>Artificial Neural Networks</u>***.
